## 전체 64x64로 변환

```python
import os
import pandas as pd
from PIL import Image
import os, glob, numpy as np
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('E:\\Senitment_data\\csv\\face.csv', index_col = 0)
dataset
```

```python
path_dir = 'E:\\Senitment_data\\dot_part_jpg\\mouth'
file_list = os.listdir(path_dir)
dataset = pd.DataFrame()
dataset['path'] = file_list
dataset
```

```python
dataset['path'] = 'E:\\Senitment_data\\dot_part_jpg\\mouth\\' + dataset['path']
dataset
```

```python
# 성별, 나이, 감정 칼럼 생성
inform = pd.read_excel('E:\\Senitment_data\\참가자정보.xlsx')
inform
```

```python
inform.columns = ['no', 'gender', 'age']
inform
```

```python
dataset['no'] = None
for i in range(0, len(dataset)) :
    dataset['no'][i] = dataset['path'][i].split('\\')[4].split('-')[1]
dataset
```

```python
dataset['no'] = pd.to_numeric(dataset['no'])
```

```python
dataset = pd.merge(dataset, inform.drop_duplicates(['no']), on="no", how="left")
dataset
```

```python
dataset['emo'] = None
for i in range(0, len(dataset)) :
    dataset['emo'][i] = dataset['path'][i].split('\\')[4].split('-')[2]
    
dataset
```

```python
for i in range(0, len(dataset)) :
    if dataset['emo'][i] == 'ang' :
        dataset['emo'][i] = 0
    elif dataset['emo'][i] == 'dis' :
        dataset['emo'][i] = 1
    elif dataset['emo'][i] == 'fea' :
        dataset['emo'][i] = 2
    elif dataset['emo'][i] == 'hap' :
        dataset['emo'][i] = 3
    elif dataset['emo'][i] == 'sad' :
        dataset['emo'][i] = 4
    elif dataset['emo'][i] == 'sur' :
        dataset['emo'][i] = 5
    elif dataset['emo'][i] == 'neu' :
        dataset['emo'][i] = 6
          
dataset
```

```python
dataset['light'] = None
for i in range(0, len(dataset)) :
    if dataset['no'][i] <= 40 :
        dataset['light'][i] = 1
    elif dataset['no'][i] > 40 :
        dataset['light'][i] = 0
dataset
```

```python
dataset.to_csv('E:\\Senitment_data\\csv\\dot_mouth.csv', sep=',', na_rep='NaN')
```

### pixel화

```python
image_w = 64
image_h = 64

data = []

for i in dataset['path']:
        img = Image.open(i)
        img = img.convert("L")
        #img = img.resize(image_w * image_h)
        img = img.resize((image_w, image_h))

        img_data = np.asarray(img)

        data.append(img_data)
        
data
```

```python
dataset['pixels'] = None

for i in range(0,len(data)):
    dataset['pixels'][i] = data[i].flatten()
    
dataset
```

```python
data = dataset['pixels']
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
labels_one_hot = to_categorical(dataset['emo'])
labels_one_hot
```

```python
y_df_train = labels_one_hot
X_df_train = data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_df_train, y_df_train,
                                                    stratify = y_df_train,
                                                    test_size=0.1, random_state=1234)
```

```python
#np.vstack stack arrays in sequence vertically (picking element row wise)
X_train = np.vstack(X_train)
X_test = np.vstack(X_test)

#Reshape X_train, y_train,X_test,y_test in desired formats
X_train = np.reshape(X_train, (X_train.shape[0],64,64,1))
# y_train = np.reshape(y_train, (y_train.shape[0],1))
X_test = np.reshape(X_test, (X_test.shape[0],64,64,1))
# y_test = np.reshape(y_test, (y_test.shape[0],1))

print("Shape of X_train and y_train is " + str(X_train.shape) +" and " + str(y_train.shape) +" respectively.")
print("Shape of X_test and y_test is " + str(X_test.shape) +" and " + str(y_test.shape) +" respectively.")
```

```python
np.save('E:\\Senitment_data\\train_test\\dot_mouth\\X_train.npy', X_train) 
np.save('E:\\Senitment_data\\train_test\\dot_mouth\\y_train.npy', y_train) 
np.save('E:\\Senitment_data\\train_test\\dot_mouth\\X_test.npy', X_test) 
np.save('E:\\Senitment_data\\train_test\\dot_mouth\\y_test.npy', y_test) 
```
